### **SQL 분류**

SQL은 크게 데이터 **조작어**, **정의어**, **제어어**로 나뉩니다.

* **데이터 정의어** : 테이블 생성 변경 제거
* **데이터 조작어** : 데이터 삽입 수정 삭제 검색
* **데이터 제어어** : 보안 위해 데이터 접근 권한 부여 혹은 취소




### **과제 1. 기본 스키마 생성**
아래 스키마를 생성해봅니다.


image.png

테이블을 검색해보고, 기존 테이블을 삭제합니다.

```
-- 테이블 검색 --
SELECT * FROM Customer;
```

```
-- 테이블 삭제 --
DROP TABLE Customer;
DROP TABLE Customer_Package;
```


새로운 테이블을 생성합니다.

```
-- 테이블 생성 -- 
create table Customer(
customer_id int not null, 
customer_name varchar(32) not null,
customer_age int,
primary key(customer_id))

create table Package(
package_id int not null,
package_name varchar(32) not null,
package_date date,
primary key(package_id))

create table Customer_Package(
cp_id int not null,
customer_id int,
package_id int,
primary key(cp_id),
foreign key(customer_id) REFERENCES Customer(customer_id),
foreign key(package_id) REFERENCES Package(package_id))
```
